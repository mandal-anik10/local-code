# Calculaton of the mass of a planet by Kepler's 3rd law:

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Observed Datas:
Ms = 1989000*10**24       #kg
AU = 149.6*10**9         #m
planet = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune', 'Pluto']
MjA = np.array([0.774196, 1.446, 2, 3.048, 10.4088, 19.7052, 38.4368, 60.22, 79])     #AU
TP = np.array([0.241, 0.6156, 1, 1.8807, 11.8565, 29.4, 84.02, 164.79, 247.92]) * 31536000     #s
M = np.array([0.055,0.815,1,0.10744,317.82,95.16,14.371,17.147,0.0022])*5.9737*10**24     #kg
G = 6.67*10**(-11)     #SI unit
```

```python
# Actual Formula between MA and TP:
# T^2 = (pi^2/2G(M+m))*A^3
# Kepler's suggestion:
# T^2 = (pi^2/2GM)*A^3
```

```python
CK = []
CA = []
CS = []
Y0 = []
for i in range(len(planet)):
    ck = (np.pi**2)/(2*G*Ms)            #(pi^2/2GM)
    ca = (TP[i]**2)/(MjA[i]*AU)**3      #(T^2/A^3)
    cs = (np.pi**2)/(2*G*(Ms+M[i]))     #(np.pi**2)/(2*G*(M+m)
    y0 = ck-ca
    CK.append(ck)
    CA.append(ca)
    CS.append(cs)
    Y0.append(y0)
```

```python
MP = []
for i in range(len(planet)):
    fc = (np.pi**2)/(2*G*Ms*Y0[i])
    mp = Ms/(fc-1)
    MP.append(mp)
    print("The calculated mass of the planet ", planet[i], 'is :', mp,'  kg')
```

```python
plt.plot(planet,CK,'r',planet,CA,'g',planet,CS,'c')
plt.title('Planet Mass Vs constant(T^2/A^3)')
plt.legend(['$Const. Suggested In Kepler,s 3rd Law$','$Actual Constant Of$ $T^2 = k*A^3$ equ.','$Const.Should To Be$'],loc = 3)
plt.show()
```
